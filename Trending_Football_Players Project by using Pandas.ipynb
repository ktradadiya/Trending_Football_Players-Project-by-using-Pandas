* Convert Value and Wage column into Numeric Type.

```python
def convert_num(y):
    try:
        if y.endswith("K")==True:
            y = y.replace('K','')
            if y.startswith("€")==True:
                y = y.replace('€','')
                try:
                    return float(y)*1000
                except:
                    return y
            else:
                pass
           
        elif y.endswith("M")==True:
            y = y.replace('M','')
            if y.startswith('€')==True:
                y = y.replace('€','')
                try:
                    return float(y)*10000000
                except:
                    return y
            else:
                pass
            
    except:
        return y
            
                    
```

```python
convert_num("€39.5M")
```

```python
col_to_transform = ['Wage','Value']

for i in col_to_transform:
    df[i] = df[i].apply(lambda x: convert_num(x))
```
